```python
# Install required packages
# !pip install gradio transformers

import gradio as gr
from transformers import pipeline

# 🔧 Load the IBM Granite model pipeline for text generation
def load_model():
    return pipeline("text-generation", model="ibm-granite/granite-3.3-2b-instruct")

pipe = load_model()

# 💬 Generate personalized financial advice
def generate_advice(user_input, user_type):
    prompt = (
        f"You are a helpful personal finance assistant. "
        f"Your tone and complexity should be adjusted for a {user_type}. "
        f"Provide personalized financial guidance on savings, taxes, investments, "
        f"budget summaries, and spending insights.\n\n"
        f"User: {user_input}\n"
        f"Assistant:"
    )
    response = pipe(prompt, max_length=512, do_sample=True, temperature=0.7)
    generated_text = response[0]['generated_text']
    reply = generated_text[len(prompt):].strip()
    return reply

# 🎨 Define Gradio interface
def create_interface():
    with gr.Blocks() as demo:
        gr.Markdown("## 💰 Personal Finance Chatbot powered by IBM Granite")

        user_type = gr.Radio(
            choices=["student", "professional"],
            label="Select your user type"
        )

        user_input = gr.Textbox(
            lines=4,
            placeholder="Ask your financial question here...",
            label="Your Question"
        )

        output = gr.Textbox(
            lines=10,
            label="Chatbot Response"
        )

        submit_btn = gr.Button("Get Advice")
        submit_btn.click(fn=generate_advice, inputs=[user_input, user_type], outputs=output)

    return demo

# 🚀 Launch the app
if __name__ == "__main__":
    create_interface().launch()
```
